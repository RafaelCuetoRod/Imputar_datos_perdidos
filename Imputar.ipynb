## **Crear copia de la data**

```python
data2=data.copy()
data3=data.copy()
data2["Excess Return"],data2["Systematic Risk.1"]=data["Excess Return"],data["Systematic Risk.1"]
```

# **ARREGLAR LA DATA**

```python
def impute_nan_meanmedian(data2, variable, value):
    if value=="median":
        val=data2[variable].median()
    if value=="mean":
        val=data2[variable].mean()
    data2[variable+"_"+value]=data2[variable].fillna(val)
    return data2
import random
def rand_float_range(start, end):
    return random.random() * (end - start) + start

def impute_nan_random(data2, variable, value=-9):
    if value!=-9:
        data2[variable+"_random_fixed"]=data2[variable].fillna(value)
    else:
        start,end=data2[variable].min(),data2[variable].max()
        data2[variable+"_random"]=data2[variable].fillna(rand_float_range(start,end))
    return data2
```

# **Se incluye un valor aleatorio en la columna de datos faltantes**

```python
impute_nan_random(data2,"Excess Return",-9)
```

## **Se incluye los valores de la media en la columna de datos faltantes**

```python
impute_nan_meanmedian(data2,"Excess Return","mean")
```

## **Se incluye los valores de la mediana en la columna de datos faltantes**

```python
impute_nan_meanmedian(data2,"Excess Return","median")
```

# **Se incluye un valor aleatorio en la columna de datos faltantes**

```python
impute_nan_random(data2,"Excess Return",0.02)
```

# **Utilizando el metodo de KNN para eliminar los datos faltantes en la columna seleccionada**

```python
from sklearn.impute import KNNImputer
# define imputer
imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
# fit on the dataset
imputer.fit(data3)
# transform the dataset
Xtrans = imputer.transform(data3)
```

```python
Xtrans
```

```python

Xtrans=pd.DataFrame(Xtrans,columns=data3.columns)
```
